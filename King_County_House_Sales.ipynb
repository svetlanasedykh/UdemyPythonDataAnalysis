# Working With Dates Exercise

```python
import pandas as pd
```

```python
# Import the "kc_house_data.csv" and parse the "date" column in a single line
df = pd.read_csv("data/kc_house_data.csv", parse_dates = ["date"])
df.head()
```

```python
df["date"].min()
```

```python
df["date"].max()
```

```python
# Calculate how many days the dataset spans
number_of_days = df["date"].max() - df["date"].min()
number_of_days.days
```

```python
#find the date that had the most house sales
#find all the homes sold on that date
```

```python
top_date = df["date"].mode()[0]
top_date
```

```python
df[df["date"] == top_date]
```

```python
# Create a pie chart showing the years that home sales took place in (2014 and 2015)
```

```python
df["date"].dt.year.value_counts().plot(kind = "pie")
```

```python
#Find all the homes sold in 2014, sorted from earliest to latest
```

```python
df["date"].dt.year.value_counts()
```

```python
df[df["date"].dt.year == 2014].sort_values(by = "date")
```

```python
#Count up the number of sales that took place in each month of the year.  Which month had the most?
```

```python
df["date"].dt.month.value_counts()
```

```python
df["date"].dt.month.mode()[0]
```

```python
# Find the homes sold between May 1st 2014 and May 1st 2015
# Create a bar plot showing the total number of sales per month in that period
# The x axis should be in calendar order (1-12)
```

```python
df[df["date"].between("2014-05-01", "2015-05-01")].sort_values(by = "date")
```

```python
sales_by_month = df[df["date"].between("2014-05-01", "2015-05-01")].sort_values(by = "date")
```

```python
sales_by_month["date"].dt.month.value_counts()
```

```python
sales_by_month["date"].dt.month.value_counts().sort_index().plot(kind = "bar")
```

```python
# Count up the number of homesales by day of the week
```

```python
df["date"].dt.dayofweek.value_counts()
```

```python
# Create a line plot showing the total number of sales by week of the year number
```

```python
df["date"].dt.isocalendar().week.value_counts().sort_index().plot()
```

```python
#Create a bar plot showing the total number of sales that took place in December, January, and February
```

```python
winter_sales = df[df["date"].dt.month.isin([12, 1, 2])]
```

```python
winter_sales["date"].dt.month.value_counts()
```

```python
winter_sales["date"].dt.month.value_counts().plot(kind = "bar")
```

```python
#Work with the homes sold between May 1st 2014 and May 1st 2015
#Within that year, find the waterfront homes that were sold
#Which quarter of that year had the most waterfront home sales?  The least?
#Create a bar plot showing the number of waterfront home sales per quarter
```

```python
sales_by_month = df[df["date"].between("2014-05-01", "2015-05-01")].sort_values(by = "date")
```

```python
waterfront = sales_by_month[sales_by_month["waterfront"]== 1]
```

```python
waterfront["date"].dt.quarter.value_counts().sort_index().plot(kind = "bar")
```
